```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
u = np.loadtxt('u.txt')
u_true = np.loadtxt('u_true.txt')
u.shape
```

```python
for j in range(0, u.shape[0], int(u.shape[0] / 10)):
    plt.plot(u[j], ls='--', c='red', alpha=1 - (j / u.shape[0]))
    plt.plot(u_true[j], c='gray', alpha=1 - (j / u.shape[0]))
plt.ylabel('u')
plt.xlabel('x step')
plt.title(f'Nt, Nx = {u.shape}')
```
