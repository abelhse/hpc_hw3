```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pathlib
out_file = pathlib.Path('./out_task1/slurm-2319056.out')
# out_file = pathlib.Path('./out_task1/slurm-2319305.out')
```

```python
res = []
with open(out_file, 'rt') as f:
    content = f.read()
i = content.find('maxAbsErr')
content = content[i:]
d = {}
for line in content.split('\n'):
    a= line.split('=')
    if len(a) == 1:
        continue
    k, v = a
    d[k] = v
# print(d)
d['x'] = np.fromstring(d['x'], sep=' ')
d['u'] = np.fromstring(d['u'], sep=' ')
d['u_true'] = np.fromstring(d['u_true'], sep=' ')
d['filename'] = str(out_file)
d['block_sizes'] = np.asarray(list(map(int, d['block_sizes'].split())))

import json
print(content)
# break
```

```python
# for j in range(0, u.shape[0], int(u.shape[0] / 10)):
j=0
plt.plot(d['x'], d['u'], marker='x', c='red', label='numerical')
plt.plot(d['x'], d['u_true'] , marker='x', c='gray', label='formula')
plt.ylabel('u')
plt.xlabel('x')
plt.legend()
plt.title(f'Nx = {d["Nx_global"]}, Nt={d["Nt"]}, T={d["T"]}, h={d["h"]}, tau={d["tau"]},\nmaxAbsErr={d["maxAbsErr"]}')
```
